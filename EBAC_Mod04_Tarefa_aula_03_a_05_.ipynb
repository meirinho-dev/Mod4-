## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# Criando dicionário com o score total de cada matéria

total_scores = {
    'Matemática': 24,
    'Português': 18,
    'Geografia': 8,
    'Inglês': 8,
    'História': 8,
    'Física': 12,
    'Química': 12
}

# Calculando o percentual de acertos em cada matéria e salvando em uma variável
df_percentual = pd.DataFrame({
    'Matemática': (df_mat['Qt_acertos'] / total_scores['Matemática']) * 100,
    'Português': (df_por['Qt_acertos'] / total_scores['Português']) * 100,
    'Geografia': (df_geo['Qt_acertos'] / total_scores['Geografia']) * 100,
    'Inglês': (df_ing['Qt_acertos'] / total_scores['Inglês']) * 100,
    'História': (df_his['Qt_acertos'] / total_scores['História']) * 100,
    'Física': (df_fis['Qt_acertos'] / total_scores['Física']) * 100,
    'Química': (df_qui['Qt_acertos'] / total_scores['Química']) * 100
})

# Visualizando as primeiras linhas
df_percentual.head()

````

````python
# Concatenando os df de cada disciplina horizontalmente
df_total_acertos = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)

# Renomeando as colunas
df_total_acertos.columns = ['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química']

# Calculando o total de acertos de cada aluno
df_total_acertos['Total_Acertos'] = df_total_acertos.sum(axis=1)

df_total_acertos.head()

````

````python
# Calculando o total máximo de acertos por matéria
total_possible_points = sum(total_scores.values())

# Calculando o percentual de acertos de cada aluno
df_total_acertos['Percentual_Geral'] = (df_total_acertos['Total_Acertos'] / total_possible_points) * 100

df_total_acertos.head()
````

````python
# Criando uma variável para os alunos com nota superior a 45
alunos_acima_corte = df_total_acertos[df_total_acertos['Total_Acertos'] > 45].shape[0]

alunos_acima_corte
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# Criando um novo DataFrame para alunos desqualificados
df_desqualificados = df_total_acertos[
    (df_total_acertos['Matemática'] == 0) |
    (df_total_acertos['Física'] == 0) |
    (df_total_acertos['Química'] == 0)
]

df_desqualificados.head()
````

````python
# Contando quantos alunos foram desqualificados
num_desqualificados = df_desqualificados.shape[0]
num_desqualificados
````

````python
# Calculando a média de História e Geografia para os alunos desqualificados
import numpy as np
media_historia = np.mean(df_desqualificados['História'])
media_geografia = np.mean(df_desqualificados['Geografia'])

# Apresentando o resultado
print(f"A média de História é: {media_historia}")
print(f"A média de Geografia é: {media_geografia}")

````

````python
# Remover os desqualificados com base no índice
df_passaram = df_total_acertos.drop(df_desqualificados.index)

# Aplicar a condição de Percentual Geral > 45
df_passaram = df_passaram[df_passaram['Percentual_Geral'] > 45]

# Visualizar as primeiras linhas do DataFrame resultante
df_passaram.head()

````
